```python
import pandas as pd
df = pd.read_csv(r"C:\Users\dell\Desktop\MLData\carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final
```

```python
X = final.drop('Sell Price($)',axis='columns')
X
```

```python
letX = X[['Mileage',	'Age(yrs)',	'Audi A5',	'BMW X5']].values
letX
```

```python
lety = final['Sell Price($)'].values
lety
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
model.score(X,y)
```

```python
model.fit(letX,lety)
model.score(letX,lety)
```
